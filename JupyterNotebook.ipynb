```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Simulación de un perfil de velocidad trapezoidal
t = np.linspace(0, 10, 500)
v_max = 50.0
t_accel = 2.0
t_decel = 8.0

# Lógica del perfil
velocidad = np.piecewise(t, 
    [t < t_accel, (t >= t_accel) & (t <= t_decel), t > t_decel],
    [lambda t: (v_max / t_accel) * t, 
     v_max, 
     lambda t: v_max - (v_max / (10 - t_decel)) * (t - t_decel)]
)

# Graficar
plt.plot(t, velocidad, color='teal', linewidth=2.5, label='Velocidad del Actuador')
plt.fill_between(t, velocidad, color='teal', alpha=0.2)
plt.title('Perfil de Velocidad Trapezoidal del Eje X')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (RPM)')
plt.legend()
plt.show()
```
